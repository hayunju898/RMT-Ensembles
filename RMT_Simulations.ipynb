## Random Matrix Theory Simulations for MATH30033

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Gaussian Orthogonal Ensemble (GOE)

```python
#generating a random GOE (Gaussian Orthogonal Ensemble) with beta = 1
def goe_matrix(size):
    M = np.random.randn(size, size) #random complex matrix with standard normal entries
    return (M + M.T) / 2 #symmetric matrix

#simulation of the eigenvalues of random matrices from the GOE
def simulate_eigenvalues_goe(matrix_size, num_simulations):
    eigenvalues = []
    for _ in range(num_simulations):
        matrix = goe_matrix(matrix_size)
        eigenvalues.extend(np.linalg.eigvalsh(matrix))
    return np.array(eigenvalues)

#plot of eigenvalue distributions
def plot_eigenvalue_distribution_goe(eigenvalues, matrix_size):
    plt.hist(eigenvalues, bins=50, density=True, alpha=0.75, color='r', label=f'Matrix Size = {matrix_size}')
    plt.xlabel('Eigenvalue')
    plt.ylabel('Density')
    plt.title('Eigenvalue Distribution (GOE)')
    plt.legend()
    plt.show()

def main():
    matrix_size = 100  #size of random matrix
    num_simulations = 1000  #number of matrices to simulate

    #simulation and plot for GOE (beta = 1)
    eigenvalues_goe = simulate_eigenvalues_goe(matrix_size, num_simulations)
    plot_eigenvalue_distribution_goe(eigenvalues_goe, matrix_size)

if __name__ == "__main__":
    main()
```

## Circular Unitary Ensemble (CUE)

```python
#generating a random unitary matrix CUE (Circular Unitary Ensemble)
def cue_matrix(size):
    Z = np.random.randn(size, size) + 1j * np.random.randn(size, size) #random complex matrix with its entries being standard normal
    Q, R = np.linalg.qr(Z)
    R_diag = np.diagonal(R)
    L = np.diag(R_diag / np.abs(R_diag))
    return np.dot(Q, L)

#simulation of eigenvalues of random unitary matrices from the CUE
def simulate_eigenvalues_cue(matrix_size, num_simulations):
    eigenvalues = []
    for _ in range(num_simulations):
        matrix = cue_matrix(matrix_size)
        eigenvalues.extend(np.linalg.eigvals(matrix))
    return np.array(eigenvalues)

#plot of eigenvalue distributions on the complex plane
def eigenvalue_distrib_cue(eigenvalues):
    plt.figure(figsize=(8, 8))
    plt.plot(np.real(eigenvalues), np.imag(eigenvalues), 'o', alpha=0.5)
    plt.xlabel('Real Part')
    plt.ylabel('Imaginary Part')
    plt.title('Eigenvalue Distribution on the Complex Plane (CUE)')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)
    plt.show()

def main():
    matrix_size = 100  #size of random matrix
    num_simulations = 1000  #total number of matrices to simulate

    #simulation and plot for CUE (unitary matrices)
    eigenvalues_cue = simulate_eigenvalues_cue(matrix_size, num_simulations)
    eigenvalue_distrib_cue(eigenvalues_cue)

if __name__ == "__main__":
    main()
```

## Circular Orthogonal Ensemble (COE)

```python
#generating a random orthogonal matrix COE (Circular Orthogonal Ensemble)
def coe_matrix(size):
    Z = np.random.randn(size, size) #random matrix with standard normal entries
    Q, R = np.linalg.qr(Z)
    R_diag = np.diagonal(R)
    L = np.diag(np.sign(R_diag))
    return np.dot(Q, L)

#simulating eigenvalues of random orthogonal matrices from the COE
def simulate_eigenvalues_coe(matrix_size, num_simulations):
    eigenvalues = []
    for _ in range(num_simulations):
        matrix = coe_matrix(matrix_size)
        eigenvalues.extend(np.linalg.eigvals(matrix))
    return np.array(eigenvalues)

#plot of the eigenvalue distributions on the complex plane
def eigenvalue_distrib_coe(eigenvalues):
    plt.figure(figsize=(8, 8))
    plt.plot(np.real(eigenvalues), np.imag(eigenvalues), 'o', alpha=0.5)
    plt.xlabel('Real Part')
    plt.ylabel('Imaginary Part')
    plt.title('Eigenvalue Distribution on the Complex Plane (COE)')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)
    plt.show()

def main():
    matrix_size = 100  #size of random matrix
    num_simulations = 1000  #number of matrices to simulate

    #simulation and plot for COE (orthogonal matrices)
    eigenvalues_coe = simulate_eigenvalues_coe(matrix_size, num_simulations)
    eigenvalue_distrib_coe(eigenvalues_coe)

if __name__ == "__main__":
    main()
```

## Wigner Semicircle

```python
#generating a random GOE (Gaussian Orthogonal Ensemble) with beta = 1
def goe_matrix(size):
    M = np.random.randn(size, size) #random complex matrix with standard normal entries
    return (M + M.T) / 2 #symmetric matrix

#simulation of the eigenvalues of random matrices from the GOE
def simulate_eigenvalues_goe(matrix_size, num_simulations):
    eigenvalues = []
    for _ in range(num_simulations):
        matrix = goe_matrix(matrix_size)
        eigenvalues.extend(np.linalg.eigvalsh(matrix))
    return np.array(eigenvalues)

#plotting eigenvalue distributions and Wigner semicircle
def plot_eigenvalue_distribution_goe(eigenvalues, matrix_size):
    #eigenvalues
    plt.hist(eigenvalues, bins=50, density=True, alpha=0.75, color='r', label='Eigenvalue Distribution')

    #Wigner semicircle law
    x = np.linspace(-2*np.sqrt(matrix_size), 2*np.sqrt(matrix_size), 1000)
    rho = (1 / (np.pi * matrix_size)) * np.sqrt(2*matrix_size - x**2)
    plt.plot(x, rho, color='b', lw=2, label='Wigner Semicircle Law')

    plt.xlabel('Eigenvalue')
    plt.ylabel('Density')
    plt.title('Eigenvalue Distribution (GOE) and Wigner Semicircle Law')
    plt.legend()
    plt.show()

def main():
    matrix_size = 100
    num_simulations = 1000
    #simulation and plot for GOE (beta = 1)
    eigenvalues_goe = simulate_eigenvalues_goe(matrix_size, num_simulations)
    plot_eigenvalue_distribution_goe(eigenvalues_goe, matrix_size)

if __name__ == "__main__":
    main()
```
